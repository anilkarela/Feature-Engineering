```python
import pandas as pd
import datetime
```

```python
df_base=datetime.datetime.today()
df_date_list = [df_base - datetime.timedelta(days=x) for x in range(0,20)]
df=pd.DataFrame(df_date_list)
df.columns=['day']
df
```

```python
# extract the week day name

df['day_of_week'] = df['day'].dt.day_name()
df.head()
```

```python
# Engineer categorical variable by ordinal number placement
dayname_map={
    'Monday':1,
    'Tuesday':2,
    'Wednesday':3,
    'Thursday':4,
    'Friday':5,
    'Saturday':6,
    'Sunday':7
}

df['day_ordinal']=df.day_of_week.map(dayname_map)
df.head(15)
```
